# Plotting with pandas

We've learned in previous lessons how to plot using pandas. The pandas plotting functions are a nice, quick way to visualize your data. Here are some reminder examples of the kinds of plots you can produce.

One of the most-used plots in epidemiology is a histogram of categorical data. This is accomplished in pandas by first calling value_counts() to tally the number of values in each category, and then calling plot() with the bar argument.

```python
data.cluster.value_counts().plot(kind='bar')
```

You could also use a pie plot, but pie plots are generally not good at conveying information and I would advice against using them in most cases. 

```python
data.sex.value_counts().plot(kind='pie')
```

The default is a plain line plot. 

As an aside, this seems like a silly plot. Why would you bother plotting the case ids, when it just results in a straight line? The answer is that it's an excellent way to check that you have all the data you need, and that it's behaving as expected. If you have missing data, or misnumbered data, this plot would show it immediately.

```python
data.id.plot()
```

Histogram of numeric variable. Use the bins argument to select the number of bars you need. The default is 10, no matter what the range of the data is.

```python
data.age.hist(bins=20)
```

Box plots can be nice for visualizing summary statistics.

```python
data.age.plot(kind='box')
```

For most of these plot types, you can overlay two or more variables by calling them both.

```python
data[['age', 'incubation_period']].plot(kind='box')
```

Here's another example of calling two variables.

```python
data[['age', 'incubation_period']].plot() #here I'm using the [[]] construct to call >1 variable.
```

You can also produce scatter and hexbin plots, but the syntax is a little different. Instead of calling both the x and y variables using the [[]] construct, you put the variable names in as x and y arguments.

```python
data.plot(kind='scatter', x='age', y='incubation_period')
```

Pandas will automatically use the dataframe's index as the x axis. If the index is a date, it will automatically arrange and format those dates. This is quite nice for epidemiology. Here I will replace the index of our data with our Date column, then plot the case ids by those dates.

```python
data.index = data.date
```

```python
data.age.plot()
```

# Plotting with seaborn

Now, you'll notice that the above plots are quite ugly. You could change the aesthetics manually (we'll learn how below in the matplotlib section), but an easier way is to import a graphics package that will do it for you. My favorite is seaborn. I'll import it below, and show you how it instantly beautifies your visualizations. Without doing any extra work, we get nicer colors and cleaner lines.

```python
import seaborn as sns
```

```python
data[['age', 'incubation_period']].plot()
```

```python
data.plot(kind='scatter', x='age', y='incubation_period')
```

Seaborn also has a lot of nice options for quickly changing the feel of your graphics. Check out the documentation for the full range, but here are a few I use regularly.

The set_context function changes the font and size of your graphics. Once you call it, that style will carry through all of the figures you makes, so you only need to call it once.

```python
sns.set_context('talk') #options are 'paper' for tiny, 'talk' for medium, and 'poster' for huge.
```

```python
data[['age', 'incubation_period']].plot()
```

```python
sns.set_context('paper')
```

```python
data[['age', 'incubation_period']].plot()
```

There are also several options for figure aesthetics.

```python
sns.set_style('white')
```

```python
data[['age', 'incubation_period']].plot()
```

The despine() option is really nice if you go with white the white style graphics. It removes the extra black lines on the top and right side of the plot.

```python
data[['age', 'incubation_period']].plot()
sns.despine()
```

There are also numerous beautiful color palettes. I'll show just one example here, but the full range can be found at:  http://stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html#palette-tutorial

If you pick a new color palette, you'll have to save it as a variable and pass it into your plot() function.

```python
colors = sns.color_palette("husl")
data[['age', 'incubation_period']].plot(color=colors)
sns.despine()
```

Seaborn also has a bunch of beautiful and useful plots. There are too many to describe here, but one indispensable piece is the FacetGrid. This function is a building block for many of the seaborn plots. It stratifies data and visualizes each stratification in a different pane. Instead of calling your data and appending the plot() function like you would with pandas, you call the function and put your plotting requirements into the function. These can be a tad unintuitive to build, so don't be afraid to play around until you get the layout you want. See below.

This first piece sets the stage by producing each pane stratified by the columns you want. Then you call the plot you set up (here I named it g) and map it to a matplotlib function.

```python
g = sns.FacetGrid(data, row='cluster', col='sex')
g.map(plt.hist, 'age')
```

You can also use the "hue" argument to add an additional dimension. What I like about these plots is you can represent several dimensions of data, and compare across groups - which we do a lot in epidemiology.

```python
g = sns.FacetGrid(data, col='sex', hue='cluster')
g.map(plt.scatter, 'age', 'incubation_period')
```

# Plotting with matplotlib

Underlying both pandas plotting and seaborn is the package matplotlib. At some point, you will need to incorporate matplotlib into your visualizations. Plots are built piece by piece, which makes them almost infintely customizable but tedious to build. Any time you want a little extra customization (e.g. moving a legend, changing tick labels) you'll need to rely on matplotlib. I'll throw out a few examples here to help orient you, but know that you'll probably have google liberally when you go to do it yourself.

Start every matplotlib call with this line. Notice it has two variables, fig and ax. This line initiates an empty plot. There are a bunch of arguments you can pass into the subplots() function, but the one I use more often is the figsize() option.

```python
fig, ax = plt.subplots(figsize=(14, 4))
```

You can (and probably will) use matplotlib in combination with another plotting package like pandas. Combine them by passing your new ax variable into your plotting call.

```python
fig, ax = plt.subplots(figsize=(14, 4))
data[['age', 'incubation_period']].plot(ax=ax)
```

Now you can get into the weeds and make the tiny changes to perfect your plot. Here I'll move the legend to the right hand side.

```python
fig, ax = plt.subplots(figsize=(14, 4))
data[['age', 'incubation_period']].plot(ax=ax)
ax.legend(loc='upper right')
```

I've also switched from using pandas to making the exact same plot using base matplotlib. This allows me to mess with the aesthetics more by changing the thickness of the lines, adding markers, and turning down the opacity. I can also stack different kinds of plots, for example a line plot + a scatter plot. Finally, I changed the words in the legend using the label=argument, and the activated the legend in the last line.

Look carefully at the code block below and see if you can match each argument with the resulting plot aesthetics.

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(data.index, data.age, marker='o', linewidth=4, label='Age')
ax.scatter(data.index, data.incubation_period, marker='*', color='green', label='Incubation Period')
ax.legend()
```

You might also want to put in axis labels, and a title. I also often truncate or extend my axis, so I've included that too - notice how the ylim now extends to 200, and the xlim now only goes to March instead of April. Matplotlib has automatically changed the x axis to accomodate the new x axis.

Finally, I moved the axis outside the window. I had to make the plot smaller in order for this to work. To be honest, I'm not sure why that's necessary, which just goes to show you that matplotlib is immensely powerful but sometimes a bit quirky. I'm sure there's a way to make it work with a larger plot, with enough googlefu.

```python
fig, ax = plt.subplots(figsize=(8, 2))
data[['age', 'incubation_period']].plot(ax=ax)

ax.set_ylabel('Value')
ax.set_xlabel('Date')
ax.set_title('Age and incubation period of cases')
ax.set_ylim(0, 200)
ax.set_xlim(min(data.index), '2015-03-01')
ax.legend(bbox_to_anchor=(1.25, .8))
sns.despine()
```

And if nothing else, you'll have to call matplotlib in order to save your figures. You can save in almost any format (pdf, png, jpg) by just choosing the right suffix.

```python
fig, ax = plt.subplots(figsize=(8, 2))
data[['age', 'incubation_period']].plot(ax=ax)
ax.set_ylabel('Value')
ax.set_xlabel('Date')
ax.set_title('Age and incubation period of cases')
ax.legend(bbox_to_anchor=(1.25, .8))
sns.despine()

fig.savefig('my_favorite_plot.png')
```
