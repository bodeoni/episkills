```python
def process_file(df):
    df.age = df.age.convert_objects(convert_numeric=True)
    df.gender = df.gender.replace(['?', 'M?'], np.nan)
    
    labels = ['0-50', '51- 99']
    labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90+']
    df['age_group'] = pd.cut(df.age, range(0, 101, 10), right=False, labels=labels)
    
    df['onset'] = df.onset.dropna().apply(convert_date)
    
    return df
```

```python
def convert_date(x):
    try:
        x = datetime.datetime.strptime(x, '%m/%d/%Y')
    except:
        x = np.nan
        
    return x
```

```python
# Figures will appear in our report. Here I put two figures side by side using the ncols argument
# in the plt.subplots() call. I also add titles to my figures. 
# Finally, I put a text cell below with commentary (preferrably automatically generated!) 
# on the figures.
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(13,6))
data.age_group.value_counts(sort=False).plot(kind='bar', ax=ax[0])
data.age.plot(kind='box', ax=ax[1])
ax[0].set_title('Histogram of ages')
ax[1].set_title('Box plot of patient ages');
```

```python
mean_male = data.age[data.gender == 'M'].median()
mean_female = data.age[data.gender == 'F'].median()
if mean_male > mean_female:
    direction = 'greater than'
elif mean_male == mean_female:
    direction = 'the same'
elif mean_male < mean_female:
    direction = 'less than'
print('The median age of male patients is {}, {} than the average age of female patients at {}.'\
      .format(mean_male, direction, mean_female))
```
